#Missing Values

```python
from pandas.core.dtypes.missing import isna_all
## Let us capture all the nan values
## First lets handle Categorical features which are missing
categorical_features_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 1 and dataset[feature].dtypes == "O"]
for feature in categorical_features_nan:
    #print("length of catergorical_features_nan: {}".format(len(categorical_features_nan)))
    #print("Categorical features with missing values are as follows: ")
    print("{}:, {}% of missing values".format(feature, np.round(dataset[feature].isnull().mean(), 4)))
    
```

```python
## Replace categorical  missing value with a new label
def replace_cat_feature_nan(dataset,categorical_features_nan):
    df = dataset.copy()
    df[categorical_features_nan] = df[categorical_features_nan].fillna("Missing")
    return df

dataset = replace_cat_feature_nan(dataset, categorical_features_nan)
dataset[categorical_features_nan].isna().sum()
```

```python
dataset[categorical_features_nan].head()
```

```python
# Now lets check for numerical variables the contains missing values
numerical_features_nan = [feature for feature in dataset.columns if 
                          dataset[feature].isnull().sum()>1 and dataset[feature].dtypes != 'O']

## We will print the numerical nan variables and percentage of missing values
for feature in numerical_features_nan:
    print("{}: {}% of missing values".format(feature, np.around(dataset[feature].isnull().mean(), 4)))                        
```

## Replacing the numerical Missing Values

```python
for feature in numerical_features_nan:
    ## We will replace by using median since there are outliers
    median_value = dataset[feature].median()
    ## create a new feature to capture nan values
    dataset[feature+'nan'] = np.where(dataset[feature].isnull(), 1, 0)
    dataset[feature].fillna(median_value, inplace=True)

dataset[numerical_features_nan].isnull().sum()
```

```python
dataset[numerical_features_nan].head()
```

```python
dataset.head(30)
```

```python
## Temporal Variables (Date Time Variables)
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature] = dataset['YrSold'] - dataset[feature]
```


#Numerical Variables

##Since the numerical variables are skewed we will perform log normal distribution



```python
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']
for feature in num_features:
    dataset[feature] = np.log(dataset[feature])
```

```python
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
categorical_features
```

```python
for feature in categorical_features:
    temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature] = np.where(dataset[feature].isin(temp_df), dataset[feature], 'Rare_var')
```

```python
dataset.head(50)
```

```python
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```
